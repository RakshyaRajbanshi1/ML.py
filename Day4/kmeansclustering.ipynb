```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X,y = make_blobs(n_samples=500, centers=3, cluster_std=0.60, random_state=0)
#visualize the data
plt.scatter(X[:,0],X[:,1],s=30)
plt.title("Generated data")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
#initialize kmean with k
kmeans = KMeans(n_clusters=3)
#fit the model to the data
kmeans.fit(X)
#predict the cluster
y_kmeans=kmeans.predict(X)
#visualize the cluster
plt.scatter(X[:,0],X[:,1], c=y_kmeans,s=30, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1], c='red', s=200, alpha=0.7,marker='X')
plt.title("kmeans")
plt.show()
```

```python
#elbow method- compute WCSS( within cluster sum of square )

WCSS_list =[]
for k in range (1,11):
    Kmeans = KMeans(n_clusters=k)
    Kmeans.fit(X)
    WCSS_list.append(Kmeans.inertia_)

#plot the elbow  curve
plt.plot(range(1,11), WCSS_list, marker=0)
plt.xlabel("K")
plt.ylabel("WCSS")
plt.title("k vs WCSS")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X,y =make_circles(n_samples=500, factor=0.5, noise=0.1, random_state=42)
plt.scatter(X[:,0],X[:,1],s=30)
plt.title("Generated data")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.175, min_samples=10)
y_dbscan = dbscan.fit_predict(X)
plt.scatter(X[:,0],X[:,1], c=y_dbscan,s=50, cmap='viridis')
plt.title("DBSCAN")
plt.show()
```

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors
# compute the k-distance plot
neighbors = NearestNeighbors(n_neighbors=10)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)

#sort and plot the distance
distances =np.sort(distances[:,9], axis=0)
plt.plot(distances)
plt.xlabel("points")
plt.ylabel("Epsilon")
plt.title("k-Distance plot")
plt.show()
```
